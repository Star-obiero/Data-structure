```python
heart_df=pd.read_csv(".\\framingham.csv")
heart_df.head()
```

```python
heart_df.shape
```

```python
heart_df.info()
```

```python
heart_df.isnull().sum()
```

```python
heart_df.duplicated().sum()
```

```python
heart_df.describe()
```

```python
heart_df.plot()
```

Filling in Education column

```python
mode_val=heart_df['education'].mode()[0]
heart_df['education']=heart_df['education'].fillna(mode_val)
```

Filling in cigas per day

```python
plt.hist(heart_df['cigsPerDay'])
plt.xlabel('cigsperday')
plt.title('Distribution of cigperday')
plt.show()
```

this will be filled by the median cos it is skeewed

```python
media_val=heart_df['cigsPerDay'].median()
heart_df['cigsPerDay']=heart_df['cigsPerDay'].fillna(media_val)
```

Filling in bpmeds. now since it is categirical data it mwill be filled with the mode value

```python
mode_val=heart_df['BPMeds'].mode()[0]
heart_df['BPMeds']=heart_df['BPMeds'].fillna(mode_val)
```

Filling in total cholestol and since it is continous we will have find the distribution first then we can fill in using the the median or mean depending on the skeewedness of the distribution

```python
plt.hist(heart_df['totChol'])
plt.xlabel('totChol')
plt.title("Distribution of totalchol")
plt.show()
```

normaldistribution

```python
mean_val=heart_df['totChol'].mean()
heart_df['totChol']=heart_df['totChol'].fillna(mean_val)
```

This process will follo for all the remaining columns as they are all continous data

Fill in missing values in heart rate

```python
plt.hist(heart_df['heartRate'])
plt.xlabel('heart rate')
plt.title('Distribution of heart rate')
plt.show()
```

normal distribution

```python
mean_val=heart_df['heartRate'].mean()
heart_df['heartRate']=heart_df['heartRate'].fillna(mean_val)
```

Filling in missing values for BMI

```python
plt.hist(heart_df['BMI'])
plt.xlabel('BMI')
plt.title('Distribution of BMI')
plt.show()
```

Normaldistribution

```python
mean_val=heart_df['BMI'].mean()
heart_df['BMI']=heart_df['BMI'].fillna(mean_val)
```

Filling in missing values in  glucose

```python
heart_df['glucose']=heart_df.groupby('diabetes')['glucose'].transform(
    lambda x:x.fillna(x.median())
)
```

```python
heart_df.info()
```

the data is now clean let us now train our model now
